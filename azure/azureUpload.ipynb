```python
import os
print(os.getcwd())
```

```python
training = '../../BraTS2021/train/'
testing = '../../BraTS2021/test/'
validation = '../../BraTS2021/val/'
```

```python
print(len(os.listdir(training)), len(os.listdir(testing)), len(os.listdir(validation)))
```

```python
def list_files(path):
    brains = os.listdir(path)
    files = []
    for brain in brains:
        brain_path = os.path.join(path, brain)
        files += [os.path.join(brain_path, file) for file in os.listdir(brain_path)]
    #print(len(files))

    return files
```

```python
training_files = list_files(training)
validation_files = list_files(validation)
testing_files = list_files(testing)

print(f'training: {len(training_files)}, validation: {len(validation_files)}, testing: {len(testing_files)}')
```

```python
all_files = training_files + testing_files + validation_files
print(len(all_files))
all_files
```

```python
datastore.upload_files(files=all_files, relative_root='../../BraTS2021/', target_path='datasets/BraTS2021/')
```
